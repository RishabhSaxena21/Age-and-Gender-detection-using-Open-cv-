VISUALIZATION TECHNIQUES USING MATPLOTLIB

```
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1,20)
y = 2 * x + 5
plt.title('Line Graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.plot(x,y)
plt.show()
```

for Same plane

```
import matplotlib.pyplot as plt

x = np.array([2,4,6,8,10])
y = x*2

plt.plot(x,y)


x1 = np.array([2,10,15,20,25])
y1 = x*2
plt.title('Line Graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.plot(x1,y1, '*-')
plt.show()
```

Fill the Area Between Two Lines

```
import matplotlib.pyplot as plt
import numpy as np

x = np.array([1, 2, 3, 4])
y = x*2

plt.plot(x, y)

x1 = [2, 4, 6, 8]
y1 = [3, 5, 7, 9]

plt.plot(x, y1, '-.')
plt.xlabel("X-axis data")
plt.ylabel("Y-axis data")
plt.title('multiple plots')

plt.fill_between(x, y, y1, color='red', alpha=1)
plt.show()
```

BAR PLOT

```
import matplotlib.pyplot as plt
import numpy as np

fruits = ['Apple','Banana','Papaya','Pomegrenate','Kiwi']
sales = [2500,500,2000,8000,12000]

plt.barh(fruits,sales, color = 'Green')
plt.title('Sale')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Histogram Plot

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = np.random.randn(100)

plt.hist(data,bins = 40, color = 'blue', edgecolor = 'black')
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Histogram')
plt.show()
```

Histogram Density Plot

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

data = np.random.randn(100)

sns.histplot(data,bins = 40, kde = True, color = 'blue', edgecolor = 'black')
plt.xlabel('values')
plt.ylabel('frequency')
plt.title('Histogram with Density Plot')
plt.show()
```
